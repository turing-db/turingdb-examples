```python
dict_df = {}
```

```python
for path_csv in glob.glob(f"{path_data}/{folder}/*.csv"):
    print(100 * '-')
    key_name = re.sub('.csv', '', os.path.basename(path_csv))
    dict_df[key_name] = pd.read_csv(path_csv)

    print(f"--- {key_name}")

    display(dict_df[key_name])

print(100 * '-')
```

```python
dict_df['sequence']
```

```python
df_sequence_modified = dict_df['sequence'].copy()
df_sequence_modified['output_name'] = df_sequence_modified['goes_into_name'].combine_first(df_sequence_modified['is_type_of_name'])
df_sequence_modified['output_id'] = df_sequence_modified['goes_into_id'].combine_first(df_sequence_modified['is_type_of_id'])
df_sequence_modified['type_link'] = np.where(df_sequence_modified['goes_into_id'].notna(), 'goes_into_id', 'is_type_of_id')
df_sequence_modified['type_link'] = df_sequence_modified['type_link'].str.replace('_id', '')
df_sequence_modified = df_sequence_modified.drop(['goes_into_name', 'goes_into_id', 'is_type_of_name', 'is_type_of_id'], axis=1)
df_sequence_modified
```

```python
G = create_graph_from_df(
    df_sequence_modified,
    source_node_col={
        'id': 'input_id',
        'displayName': 'input_name'
        #'type': ''
    },
    target_node_col={
        'id': 'output_id',
        'displayName': 'output_name'
    },
    node_attributes_df=dict_df['inputs'],
    node_attributes_key_col='input_id',
    edge_type_col='type_link'
)
print(f"Resulting graph : {G}")
```

```python
n_first = 10

print("NODES :")
for node in list(G.nodes(data=True))[:n_first]:
    print(node[0])
    for key, val in node[1].items():
        print(f"   {key} : {val}")
    
    print()
print()

print("EDGES :")
for edge in list(G.edges(data=True))[:n_first]:
    print(edge)
```

```python
graph_name = "supply_chain_eto-chip-explorer2"
graph_name = re.sub('-', '_', graph_name)
```

```python
from turingdb import TuringDB

# Create TuringDB client
client = TuringDB(host="http://localhost:6666")

# Create a new graph
client.query(f"CREATE GRAPH {graph_name}")
client.set_graph(graph_name)

# Create a new change on the graph
change = client.query("CHANGE NEW")["Change ID"][0]
print(f"--> change n°{change}")

# Checkout into the change
client.checkout(change=change)

# Build CREATE command from networkx object
create_command = build_create_command_from_networkx(G)
#print(create_command)

# Run CREATE command
client.query(create_command)

# Commit the change
client.query("COMMIT")
client.query("CHANGE SUBMIT")
```
