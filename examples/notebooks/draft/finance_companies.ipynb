```python
G = create_graph_from_df(
    df,
    directed=True,  # Corporate ownership is directional
    source_node_col={
        'id': 'parent_company',
        'displayName': 'parent_company', 
        'type': 'Company'
    },
    target_node_col={
        'id': 'subsidiary_company',
        'displayName': 'subsidiary_company',
        'type': 'Company'
    },
    attributes_edges=[
        'ownership_percentage', 
        'relationship_type', 
        'jurisdiction', 
        'registration_date', 
        'source',
        'description'
    ],
    edge_type_col='relationship_type'
)
print(f"Resulting graph : {G}")
```
