```python
from rdflib import Graph as RDFGraph
```

```python
path_rdf = "./wine.rdf"
```

```python
# RDF graph loading
#path = input("Path or URI of the RDF graph to load: ")
rg = RDFGraph()
rg.parse(path_rdf) # , format='turtle')
print("rdflib Graph loaded successfully with {} triples".format(len(rg)))
```
